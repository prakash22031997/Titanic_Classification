```python
dataset=pd.read_csv('titanic.csv')
```

```python
print(dataset.shape)
```

```python
income_set = set(dataset['Sex'])
dataset['Sex'] = dataset['Sex'].map({'female': 0 , 'male': 1}).astype(int)
dataset.head()
```

```python
dataset=dataset.drop('Cabin',axis='columns')
dataset=dataset.drop('Embarked',axis='columns')
dataset=dataset.drop('Name',axis='columns')
dataset=dataset.drop('Ticket',axis='columns')
dataset=dataset.drop('PassengerId',axis='columns')
dataset=dataset.drop('SibSp',axis='columns')
dataset=dataset.drop('Parch',axis='columns')
```

```python
dataset.head()
```

```python
X =dataset.drop('Survived', axis = 'columns')
```

```python
X
```

```python
Y = dataset.Survived
```

```python
Y
```

```python
X.columns[X.isnull().any()]
```

```python
X.Age = X.Age.fillna(X.Age.mean())
```

```python
X.columns[X.isna().any()]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X, Y , test_size = 0.20 , random_state = 0)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model= GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
pclassNo =int(input("Enter Person's Pclass number: "))
gender = int(input("Enter Person's Gender: "))
age = int(input("Enter Person's Age: "))
fare = float(input("Enter Person's fare: "))
person = [[pclassNo,gender,age,fare]]
result = model.predict(person)
print(result)


if result == 1:
  print("Person might be Survived")
else:
    print("Person might not be survived")
```

Prediction for all datasets


```python
y_predict = model.predict(X_test)
print(np.column_stack((y_predict,y_test)))
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy of the Model: {0}%".format(accuracy_score(y_test,y_predict)*100))
```
